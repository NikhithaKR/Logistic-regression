```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("bank-full.csv", sep=';')
```

```python
data
```

# EDA

```python
data.info()
```

```python
data1=pd.get_dummies(data,columns=['job','marital','education','contact','poutcome','month'])
data1
```

```python
data1.info()
```

```python
data1['default'] = np.where(data1['default'].str.contains("yes"), 1, 0)
data1['housing'] = np.where(data1['housing'].str.contains("yes"), 1, 0)
data1['loan'] = np.where(data1['loan'].str.contains("yes"), 1, 0)
data1['y'] = np.where(data1['y'].str.contains("yes"), 1, 0)
data1
```

```python
data.info()
```

```python
a=pd.concat([data1.iloc[:,0:10],data1.iloc[:,11:]],axis=1)
b=data1.iloc[:,10]
```

```python
plt.figure(figsize=(10,6))
b.value_counts().plot(kind="bar")
```

```python
plt.figure(figsize=(10,6))
b.value_counts().plot(kind="pie")
```

splitting the data set

```python
x_train,x_test,y_train,y_test = train_test_split(a,b,
                                   random_state=10, 
                                   test_size=0.25, 
                                   shuffle=True)
```

```python
x_train
```

```python
y_train.head()
```

```python
x_test
```

```python
b.value_counts()
```

```python
classifier=LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

# visualize the results calculating AUC score of logistic regression using ROC curve

```python
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred)

# Plot the ROC curve
plt.plot(fpr, tpr)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.show()

# Calculating the area under the ROC curve
auc = roc_auc_score(y_test, y_pred)
print("AUC:", auc)
```
